```python
# ============================================================
# PROJECT 3: AI-POWERED TECH STACK RECOMMENDER
# Developer: Muhammad Aamir
# Internship: DecodeLabs - Batch 2026
# Description: Content-Based Filtering using TF-IDF
#              and Cosine Similarity
# ============================================================

import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import warnings
warnings.filterwarnings('ignore')

print("=" * 50)
print("   AI Tech Stack Recommender - DecodeLabs")
print("   Developer: Muhammad Aamir | Batch 2026")
print("=" * 50)
print("✅ All Libraries Loaded Successfully!")
```

```python
# ============================================================
# KNOWLEDGE BASE: Job Roles and  Required Skills
# ============================================================

data = {
    'job_role': [
        'Data Scientist',
        'Machine Learning Engineer',
        'Backend Developer',
        'Frontend Developer',
        'DevOps Engineer',
        'Data Analyst',
        'AI Engineer',
        'Cloud Architect',
        'Cybersecurity Analyst',
        'Full Stack Developer'
    ],
    'skills': [
        'python sql machine learning data analysis statistics pandas numpy',
        'python tensorflow pytorch deep learning neural networks algorithms',
        'python java sql apis rest django flask databases nodejs',
        'html css javascript react vuejs typescript ui ux design',
        'aws docker kubernetes linux git ci cd automation cloud',
        'sql excel python tableau power bi statistics data visualization',
        'python machine learning nlp computer vision tensorflow algorithms',
        'aws azure google cloud networking infrastructure automation devops',
        'networking security linux ethical hacking python firewall encryption',
        'html css javascript python react nodejs sql databases apis'
    ]
}

df = pd.DataFrame(data)

print("✅ Knowledge Base Ready!")
print(f"📊 Total Job Roles: {len(df)}")
print("\n📋 Available Job Roles:")
for i, role in enumerate(df['job_role'], 1):
    print(f"   {i}. {role}")
```

```python
# ============================================================
# USER INPUT: Minimum 3 Skills Required
# ============================================================

print("=" * 50)
print("   WELCOME TO AI TECH STACK RECOMMENDER")
print("=" * 50)
print("\n💡 Enter your top 3 skills to get job recommendations")
print("   Example: python, machine learning, sql\n")

skill1 = input("🔹 Enter Skill 1: ").strip().lower()
skill2 = input("🔹 Enter Skill 2: ").strip().lower()
skill3 = input("🔹 Enter Skill 3: ").strip().lower()

# User profile banana
user_skills = f"{skill1} {skill2} {skill3}"

print("\n✅ Your Skill Profile:")
print(f"   → {skill1.title()}")
print(f"   → {skill2.title()}")
print(f"   → {skill3.title()}")
```

```python
# ============================================================
# PROCESS: TF-IDF Vectorization + Cosine Similarity
# ============================================================

# Step 1: TF-IDF Vectorizer
vectorizer = TfidfVectorizer()

# Step 2: Job roles ko vectors mein convert karo
job_vectors = vectorizer.fit_transform(df['skills'])

# Step 3: User skills ko vector mein convert karo
user_vector = vectorizer.transform([user_skills])

# Step 4: Cosine Similarity calculate karo
similarity_scores = cosine_similarity(user_vector, job_vectors)[0]

# Step 5: Scores DataFrame mein add karo
df['match_score'] = similarity_scores

# Step 6: Score ko percentage mein convert karo
df['match_percentage'] = (df['match_score'] * 100).round(2)

print("✅ Similarity Analysis Complete!")
```

```python
# ============================================================
# OUTPUT: Top 3 Job Role Recommendations
# ============================================================

# Sort karo highest score se
top_recommendations = df.sort_values(
    by='match_score',
    ascending=False
).head(3).reset_index(drop=True)

print("\n" + "=" * 50)
print("   🏆 YOUR TOP JOB RECOMMENDATIONS")
print("=" * 50)
print(f"\n👤 Your Skills: {skill1.title()}, {skill2.title()}, {skill3.title()}")
print("\n📌 Best Matching Career Paths:\n")

medals = ["🥇", "🥈", "🥉"]

for i, row in top_recommendations.iterrows():
    print(f"{medals[i]} Rank {i+1}: {row['job_role']}")
    print(f"   Match Score: {row['match_percentage']}%")
    print(f"   Required Skills: {row['skills']}")
    print()

print("=" * 50)
print("✅ Recommendation Complete! - DecodeLabs 2026")
print("=" * 50)
```
